```python
import matplotlib.pyplot as plt

import numpy as np
```

```python

def generate_sine_wave(lower_bound, upper_bound, num_points=128):
    # Create an array of evenly spaced x-values
    x = np.linspace(0, 2 * np.pi, num_points)
    
    # Compute the corresponding y-values for a full sine wave
    y = np.sin(x)
    
    # Scale and shift the sine wave to fit within the specified bounds
    # y has bounds [-1, 1]. add 1 to y to work off a base that is [2, 0]
    y = (y + 1) * (upper_bound - lower_bound) / 2 + lower_bound

    # Format the y-values as floats
    y = np.round(y).astype(int)
    
    return y

# Example usage:
lower_bound = 0
upper_bound = 1000
y_values = generate_sine_wave(lower_bound, upper_bound)

print([y for y in y_values])
```

```python

def generate_sawtooth_wave(lower_bound, upper_bound, num_points=128):
    # Create an array of evenly spaced x-values
    x = np.linspace(0, 2 * np.pi, num_points)

    y = x


    y[y >= np.pi] -= 2* np.pi

    
    y = (y) * ((upper_bound-lower_bound)/(2* np.pi)) + (upper_bound - lower_bound) / 2 

    y = np.round(y).astype(int)

    return y

# Example usage:
lower_bound = 0
upper_bound = 1000
y_values = generate_sawtooth_wave(lower_bound, upper_bound)

print([y for y in y_values])

```

```python
def generate_triangle_wave(lower_bound, upper_bound, num_points=128):
    # Create an array of evenly spaced x-values
    x = np.linspace(0, 2 * np.pi, num_points)

    y = x * 2

    # Subtract 2 * (y - np.pi) from y for y >= pi
    y[y >= np.pi] -= 2 * (y[y > np.pi] - np.pi)

    # Set values less than -pi to 5
    y[y < -np.pi] += 2 * (-np.pi - y[y < -np.pi])

    
    y = (y) * ((upper_bound-lower_bound)/np.pi)

    y /= 2

    y += (upper_bound-lower_bound) / 2

    y = np.round(y).astype(int)

    return y

# Example usage:
lower_bound = 0
upper_bound = 1000
y_values = generate_triangle_wave(lower_bound, upper_bound)

print([y for y in y_values])
```

```python
# Example usage:
lower_bound = 0
upper_bound = 1000

def show_graph(graph):

	y_values = eval(f"generate_{graph}_wave(lower_bound, upper_bound)")

	y_values = [y for y in y_values]

	# Plot the line graph
	plt.plot(y_values)

	# Customize the plot
	plt.xlabel("X-axis Label")
	plt.ylabel("Y-axis Label")
	plt.title("Line Graph")

	# Display the graph
	plt.show()

show_graph('sine')
show_graph('triangle')
show_graph('sawtooth')
```
